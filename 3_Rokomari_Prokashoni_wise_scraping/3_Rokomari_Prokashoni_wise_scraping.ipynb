### Importing libraries

```python
from requests_html import HTMLSession
import time
session = HTMLSession()
```

### Saving in a text file

```python
def storing_in_a_file(page_no,book_names,authors,categories,ratings,reviews,original_prices,sell_prices,stocks,summaries,author_details,book_specification_lists):
  print(len(book_names))
  print(len(authors))
  print(len(categories))
  print(len(ratings))
  print(len(reviews))
  print(len(original_prices))
  print(len(sell_prices))
  print(len(stocks))
  print(len(summaries))
  print(len(author_details))
  print(len(book_specification_lists))
  n=len(book_names)

  with open(f"batighor_prokashoni_page_{page_no}.txt", "w") as f:
          f.write("No, book_names, authors, categories, ratings, reviews, original_prices, sell_prices, stocks, summaries, author_details, book_specification_lists\n")
          for n, bn,a,c,ra,re,op,sp,s,su,ad,bsl  in zip(range(1,len(book_names)+1),book_names,authors,categories,ratings,reviews,original_prices,sell_prices,stocks,summaries,author_details,book_specification_lists):
              f.write(f"{n},{bn},{a},{c},{ra},{re},{op},{sp},{s},{su},{ad},{bsl}")
              f.write('\n')
```

### Function for getting all book info of one page

```python
def rokomari_book_scraping(book_links,page_no):
    book_names = []
    authors = []
    categories = []
    ratings = []
    reviews = []
    original_prices = []
    sell_prices = []
    stocks = []
    summaries  = []
    author_details = [] 
    book_specification_lists = []
    for i in range (0, len(book_links)):
      print(book_links[i])
      r=session.get(book_links[i])  
      book_name=r.html.find("#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__header > h1")[0]
      author= r.html.find("#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__content > p.details-book-info__content-author > a")[0]
      category = r.html.find("#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__content > div.details-book-info__content-category.d-flex.align-items-center > a")[0]
      rating=r.html.find("#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__content > div.details-book-info__content-rating > span")
      try:
        book_rating=rating[0].text
      except:
        book_rating="No rating"
      review=r.html.find("#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__content > div.details-book-info__content-rating > a")
      try:
        book_review = review[0].text
      except:
        book_review = "No reviews"
      original_price=r.html.find('#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__content > div.details-book-info__content-book-price > strike')[0]
      sell_price=r.html.find("#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__content > div.details-book-info__content-book-price > span.sell-price")[0]
      book_stock="#details-page > section.details-book-section > div > div.col-9 > div > div > div.col.details-book-main-info.align-self-center > div.details-book-main-info__content > div:nth-child(6) > figure"
      try:
        stock = r.html.find(book_stock)[0].text
      except:
        stock = "-"
      summary=r.html.find('#js--summary-description')[0]
      author_detail=r.html.find('#js--author-description')[0]
      book_specification = r.html.find("#book-additional-specification")[0].text
      book_specification_list = book_specification.split("\n")

      book_names.append(book_name.text)
      authors.append(author.text)
      categories.append(category.text)
      ratings.append(book_rating)
      reviews.append(book_review)
      original_prices.append(original_price.text)
      sell_prices.append(sell_price.text)
      stocks.append(stock)
      summaries.append(summary.text)
      author_details.append(author_detail.text)
      book_specification_lists.append(book_specification_list)

    storing_in_a_file(page_no,book_names,authors,categories,ratings,reviews,original_prices,sell_prices,stocks,summaries,author_details,book_specification_lists)
  
```

### Function of getting all anchor tag of one page

```python
def one_pages_all_books(page_no):
    books_url=f'https://www.rokomari.com/book/publisher/572/batighor-prokashoni?ref=mm_p12&page={page_no}'
    req=session.get(books_url)
    book_details_link = req.html.find(".book-list-wrapper>a")
    links=[]
    book_links=[]
    for i in range(0,len(book_details_link)):
      s=book_details_link[i].attrs["href"]
      listed=s.split()
      [links.append(x) for x in listed if x not in links]
      book_url='https://www.rokomari.com'+links[i]
      book_links.append(book_url)

    start = time.time()
    rokomari_book_scraping(book_links,page_no)
    end=time.time()
    print(f"Time required ====={end-start}======")
```

```python
batighor_prokashoni_total_page=5
for i in range (1,batighor_prokashoni_total_page+1):
    one_pages_all_books(i)
```
